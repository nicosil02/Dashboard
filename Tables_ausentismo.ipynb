# Table Ausentismo

```python
import pandas as pd
data=pd.read_csv("elecciones_limpio.csv")
```

```python
# Filtrar las filas donde 'Año' es igual a 2022
data = data[data["Año"] == 2022]

data
```

```python
data = data.drop_duplicates()
```

```python
# Crear una copia del DataFrame excluyendo la columna "Electores" para identificar duplicados
data= data.drop(columns=["Electores"])

# Identificar los índices de las filas duplicadas
duplicados = data.duplicated(keep='first')

# Eliminar las filas duplicadas basándonos en todas las columnas excepto "Electores"
data = data[~duplicados]
```

```python
# Crear la variable nivel_ausentismo con nuevas reglas
def clasificar_ausentismo(valor):
    if valor > 0.30:
        return "3_Alto"
    elif valor >= 0.20:
        return "2_Alerta"
    else:
        return "1_Bajo"

data['nivel_ausentismo'] = data['P_ausentismo'].apply(clasificar_ausentismo)

# Verificar los niveles creados
print(data['nivel_ausentismo'].value_counts())
```

```python
# Ver valores únicos en la columna 'nivel_ausentismo'
print(data['nivel_ausentismo'].unique())

# Ver distribución de datos dentro de cada nivel
for nivel in data['nivel_ausentismo'].unique():
    print(f"\nValores para el nivel {nivel}:")
    print(data[data['nivel_ausentismo'] == nivel]['P_ausentismo'].describe())
```
