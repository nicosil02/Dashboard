```python
elecciones = pd.read_csv("elecciones_limpio.csv")
```

```python
elecciones
```

```python
renamu = pd.read_csv("renamu18_22.csv")
```

```python
# Renombrar 'Departamento' a 'Region' en renamu
renamu = renamu.rename(columns={'Departamento': 'Region'})

# Verificar los cambios
print(renamu.columns)
```

```python
# Actualizar las claves de unión
columnas_clave = ['Año', 'Ubigeo', 'Region', 'Provincia', 'Distrito']

# Convertir las claves a string para evitar problemas de tipo
for columna in columnas_clave:
    renamu[columna] = renamu[columna].astype(str)
    elecciones[columna] = elecciones[columna].astype(str)

# Realizar la unión
elecciones_con_utilidad = pd.merge(
    elecciones,
    renamu[['Año', 'Ubigeo', 'Region', 'Provincia', 'Distrito', 'Utilidad']],
    on=columnas_clave,
    how='left'
)

# Verificar el resultado
print(elecciones_con_utilidad.head())
```

```python
elecciones_con_utilidad
```

```python
elecciones_18_22 = elecciones_con_utilidad.dropna()
```

```python
elecciones_18_22.to_csv("elecciones_18_22.csv", index=False, encoding="utf-8")
```
