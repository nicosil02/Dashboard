# Limpiamos la data

```python
# Mantener solo las columnas necesarias y renombrar la variable P301A a edu
educa2022Q = educa2022[['AÑO', 'UBIGEO', 'P301A']]

# Mostrar las primeras filas del DataFrame modificado
print(educa2022Q.head())
```

```python
educa2022Q['edu'] = educa2022Q['P301A'].replace({
    1: 1,  # Sin nivel
    2: 2, 3: 2, 4: 2, 12: 2,  # Básica
    5: 3, 6: 3,  # Intermedia
    7: 4, 8: 4, 9: 4, 10: 4, 11: 4,  # Superior
    99: None  # Missing value
})

educa2022Q = educa2022Q.drop(columns=['P301A'])
```

```python
# Mantener solo las columnas necesarias
empleo2022Q = empleo2022[['AÑO', 'UBIGEO', 'P524A1']].rename(columns={'P524A': 'ingresos'})

# Mostrar las primeras filas del DataFrame modificado
print(empleo2022Q.head())
```

```python
# Mantener solo las columnas necesarias
goberna2022aQ = goberna2022a[['AÑO', 'UBIGEO', 'P1$01', 'P1$02', 'P1$03', 'P1$12', 'P1$11', 'P1$09', 'P1$08', 'P3', 'P7']]

# Mostrar las primeras filas del DataFrame modificado
print(goberna2022aQ.head())
```

```python
# Mantener solo las columnas necesarias
salud2022Q = salud2022[['AÑO', 'UBIGEO', 'P401C', 'P401H1', 'P401H2', 'P401H3', 'P401H4', 'P401H5', 'P401H6']]

# Mostrar las primeras filas del DataFrame modificado
print(salud2022Q.head())
```

```python
# Asegurarse de que las columnas sean numéricas
cols_to_sum = ['P401H1', 'P401H2', 'P401H3', 'P401H4', 'P401H5', 'P401H6']
salud2022Q[cols_to_sum] = salud2022Q[cols_to_sum].apply(pd.to_numeric, errors='coerce')

# Sumar todas las variables para cada fila
salud2022Q['Indice_Limitaciones_Raw'] = salud2022Q[cols_to_sum].sum(axis=1)

# Normalizar el índice sobre 100, con un máximo teórico de 12
salud2022Q['Indice_Limitaciones'] = (salud2022Q['Indice_Limitaciones_Raw'] / 12) * 100

# Calcular el valor mínimo y máximo del índice
min_val = salud2022Q['Indice_Limitaciones'].min()
max_val = salud2022Q['Indice_Limitaciones'].max()

# Imprimir los resultados
print(f"Valor mínimo del índice: {min_val}")
print(f"Valor máximo del índice: {max_val}")

# Mostrar las primeras filas para verificar
print(salud2022Q[['Indice_Limitaciones_Raw', 'Indice_Limitaciones']].head())
```

```python
salud2022Q=salud2022Q.drop(columns=['P401H1','P401H2','P401H3','P401H4','P401H5','P401H6','Indice_Limitaciones_Raw'])
```

```python
# Recodificar la variable P110A1: 1 -> 1 (sí), 2 -> 0 (no)
vivienda2022Q['P110A1'] = vivienda2022Q['P110A1'].replace({1: 1, 2: 0})

# Verificar los cambios
print(vivienda2022Q[['P110A1']].head())
```

```python
# Mantener solo las columnas necesarias
vivienda2022Q = vivienda2022[['AÑO', 'UBIGEO', 'P110A1',  'P111A', 'P1121']]

# Mostrar las primeras filas del DataFrame modificado
print(vivienda2022Q.head())
```

```python
vivienda2022Q['saneamiento'] = vivienda2022Q['P111A'].replace({
    1: 1,2:1,  #Si tiene
    3:0,4:0,5:0,6:0,7:0,8:0,9:0,  #no tiene
})

vivienda2022Q = vivienda2022Q.drop(columns=['P111A'])
```
